```python
import pandas as pd
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.ensemble import AdaBoostClassifier
from sklearn.tree import DecisionTreeClassifier
```

```python
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/resampled_SMOTEENN.csv')
print("Dataset shape:", df.shape)
print("First 5 rows:\n", df.head())
print("Class distribution:\n", df.iloc[:, -1].value_counts())
```

```python
X = df.iloc[:, :-1]
y = df.iloc[:, -1]
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42, stratify=y
)
```

```python
adaboost_model = AdaBoostClassifier(random_state=42)
```

```python
adaboost_model.fit(X_train, y_train)
```

```python
y_pred_ada = adaboost_model.predict(X_test)
```

```python
accuracy_ada = accuracy_score(y_test, y_pred_ada)
print("\n--- AdaBoost Results ---")
print(f"AdaBoost Accuracy: {accuracy_ada}")
print("\nClassification Report:\n", classification_report(y_test, y_pred_ada))
print("\nConfusion Matrix:\n", confusion_matrix(y_test, y_pred_ada))
```

```python
param_grid_ada = {
    'n_estimators': [50, 100, 200],
    'learning_rate': [0.01, 0.1, 1],
    'base_estimator': [DecisionTreeClassifier(max_depth=1), DecisionTreeClassifier(max_depth=2)]  # Example: varying the complexity of the 'weak learner'
}
```

```python
ada = AdaBoostClassifier(random_state=42)
```

```python
grid_search_ada = GridSearchCV(estimator=ada,
                           param_grid=param_grid_ada,
                           cv=10,  # 10-fold cross-validation
                           scoring='accuracy')
```

```python
grid_search_ada.fit(X_train, y_train)
```
